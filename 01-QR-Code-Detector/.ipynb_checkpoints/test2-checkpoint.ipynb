# <font style = "color:rgb(50,120,229)">QR Assigment </font>

The **QRCodeDetector** in OpenCV is a class specifically designed for detecting and decoding QR codes in images. It simplifies the process of identifying QR codes, extracting their bounding boxes, and retrieving the encoded data.

## Key Features:
1. Detection: Identifies the presence and location of a QR code in an image.
2. Decoding: Extracts and interprets the data encoded in the QR code.
3. Bounding Box: Provides the coordinates of the QR code's boundaries in the image.

## Common Methods:
1. detectAndDecode(image):
- Detects and decodes the QR code in one step.
- Returns the decoded text, bounding box, and additional information.

Example:

```python
import cv2

# Initialize QRCodeDetector
detector = cv2.QRCodeDetector()

# Load an image
image = cv2.imread( "IDCard-Satya.png")
img=image[:,:,0:3]
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect and decode
data, bbox, _ = detector.detectAndDecode(img_gray)
if data:
    print("QR Code Data:", data)
else:
    print("No QR Code detected.")
print(bbox.shape)
```

2. detect(image):
- Detects the QR code's bounding box but does not decode it.

3. decode(image, points):
- Decodes the QR code if the points (bounding box) are provided.

## Applications:
- Scanning QR codes in images or real-time video streams.
- Extracting embedded information such as URLs, text, or other encoded data.

The QRCodeDetector is a powerful and efficient tool for working with QR codes using OpenCV.

# <font style = "color:rgb(50,120,229)">Step 2: Detect QR Code in the Image</font>

```python
# Create a QRCodeDetector Object
# Variable name should be qrDecoder

###
qrDecoder = cv2.QRCodeDetector()
import numpy as np
###

# Detect QR Code in the Image
# Output should be stored in
# opencvData, bbox, rectifiedImage
# in the same order

###
opencvData, bbox, rectifiedImage = qrDecoder.detectAndDecode(image_gray)
if opencvData:
    print(f"Decoded Data: {opencvData}")
    print(f"Bounding Box: {bbox}")
    if rectifiedImage is not None:
        rectifiedImage = rectifiedImage.astype('uint8')  # Convert rectified image to uint8
    #    cv2.imshow("Rectified QR Code", rectifiedImage)  # Display the rectified QR code
    #    cv2.waitKey(0)
    #    cv2.destroyAllWindows()

bbox=bbox.astype(np.int8)
# Check if a QR Code has been detected
if opencvData != None:
    print("QR Code Detected")
else:
    print("QR Code NOT Detected")
```

#  <font style = "color:rgb(50,120,229)">Step 3: Draw bounding box around the detected QR Code </font>

```python
n = len(bbox)

# Draw the bounding box

if bbox is not None:
        for i in range(n):
            # Draw lines between the corners
            point1 = tuple(bbox[i][0])
            point2 = tuple(bbox[(i + 1) % len(bbox)][0])
            cv2.line(image_gray, point1, point2, (255, 0, 0), 2)
# Display results
plt.imshow(image_gray)
#display(image_gray,bbox)
```
